### Load features

```python
import os
from funcs.prepare_features import prepare_features
```

```python
path_to_features = '../data/04_feature/'
path_to_target = '../data/02_intermediate/'

features_list = os.listdir(path_to_features)
features_list = [i for i in features_list if not i.startswith('.')]
features_list
```

```python
train_df, test_df = prepare_features(path_to_features, features_list, path_to_target)
```

```python
train_df
```

```python
for column in train_df.columns:
    
        if 'share' in column or 'count' in column or 'nunique' in column:
            train_df[column].fillna(0, inplace=True)
            test_df[column].fillna(0, inplace=True)
```

### Train model

```python
# LightAutoML presets, task and report generation
import pandas as pd
import tqdm
import lightautoml
from lightautoml.automl.presets.tabular_presets import TabularUtilizedAutoML
from lightautoml.tasks import Task
```

```python
N_THREADS = 8
N_FOLDS = 5
RANDOM_STATE = 4224
TEST_SIZE = 0.2
TIMEOUT = 7200

task = Task('binary', metric='auc')
roles = {'target': 'target', 'id': 'PROVIDER_NPI'}

automl = TabularUtilizedAutoML(
    task = task, 
    timeout = TIMEOUT,
    cpu_limit = N_THREADS,
    general_params = {'use_algos': ['linear_l2', 'lgb_tuned']},
    reader_params = {'n_jobs': N_THREADS, 'cv': N_FOLDS, 'random_state': RANDOM_STATE},
)
```

```python
oof_pred = automl.fit_predict(train_df, roles = roles)
```

### Explore model results

```python
from sklearn.metrics import roc_auc_score
print('OOF score: {}'.format(roc_auc_score(train_df.target.values, oof_pred.data[:, 0])))
```

```python
# Fast feature importances calculation
fast_fi = automl.get_feature_scores('fast')
fast_fi = fast_fi[~fast_fi.Feature.str.startswith('Lvl_0_') & fast_fi.Importance > 0]
print(f'{fast_fi.shape[0]} non-null features left.')
print('Top-20:')
fast_fi.head(20).style.bar()
```

```python
# save model
import joblib
joblib.dump(automl, './models/fri_lgb_tuned_darkhan_automl.sav ')
```
